```python
import numpy as np
import re
```

```python
def redpr(str):
    print('\x1b[31m' + str + '\x1b[0m')
    
def greenpr(str):
    print('\x1b[32m' + str + '\x1b[0m')
    
def colorpr2(str0, str1, col1, str2, col2):
    print(str0 + '\x1b[' + col1 + 'm' + str1.rjust(8) + '\x1b[0m', '\x1b[' + col2 + 'm' + str2.rjust(16) + '\x1b[0m')
```

```python
week = 12
lastweek = week - 1
wk = str(week)
lwk = str(lastweek)

with open('Woche' + lwk + '/Spielplan_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Spielplan = list(f)

with open('Woche' + lwk + '/Tabelle_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Tabelle = list(f)
    
with open('Woche' + lwk + '/TabelleHeim_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    TabelleHeim = list(f)
    
with open('Woche' + lwk + '/TabelleAusw_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    TabelleAusw = list(f)
    
with open('Woche' + lwk + '/TabelleForm_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    TabelleForm = list(f)
    
with open('Woche' + lwk + '/Torschüsse_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Torschüsse = list(f)
    
with open('Woche' + lwk + '/Passquote_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Passquote = list(f)
    
with open('Woche' + lwk + '/Laufdistanz_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Laufdistanz = list(f)
    
with open('Woche' + lwk + '/Zweikämpfe_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Zweikämpfe = list(f)
    
with open('Woche' + lwk + '/IntensivLauf_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    IntensivLauf = list(f)
    
with open('Woche' + lwk + '/Eckbälle_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Eckbälle = list(f)
```

```python
Tab = []
TabHeim = []
TabAusw = []
TabForm = []
for i in range(18):
    Tab.append(re.sub('\t', ' ', Tabelle[i]).rstrip().lstrip())
    TabHeim.append(re.sub('\t', ' ', TabelleHeim[i]).rstrip().lstrip())
    TabAusw.append(re.sub('\t', ' ', TabelleAusw[i]).rstrip().lstrip())
    TabForm.append(re.sub('\t', ' ', TabelleForm[i]).rstrip().lstrip())
    Eckbälle[i] = re.sub('Bayern München', 'FC Bayern München', Eckbälle[i])
    Eckbälle[i] = re.sub('Bor. Mönchengladbach', 'Borussia Mönchengladbach', Eckbälle[i])
    Eckbälle[i] = re.sub('1899 Hoffenheim', 'TSG 1899 Hoffenheim', Eckbälle[i])
    Eckbälle[i] = re.sub('Bayer Leverkusen', 'Bayer 04 Leverkusen', Eckbälle[i])
    Eckbälle[i] = re.sub('Werder Bremen', 'SV Werder Bremen', Eckbälle[i])
    Eckbälle[i] = re.sub('Fortuna Düsseldorf', 'Fortuna Düsseldorf 1895 e.V.', Eckbälle[i])
    Eckbälle[i] = re.sub('SC Freiburg', 'Sport-Club Freiburg', Eckbälle[i])
```

```python
team_list = []
data_dict = {}

for i in range(len(Passquote)):
    
    # Passquote:
    team = Passquote[i].lstrip().rstrip()[3:-3]
    team_list.append(team)
    
    rnk = int(Passquote[i][0:2].rstrip())
    
    data_dict[team] = {}
    data_dict[team]['Passquote'] = rnk
    
team_list = sorted(team_list)

for i in range(len(Passquote)):
    
    # Torschüsse:
    end = Torschüsse[i].rstrip()[::-1].index(' ')
    end += 1
    team = Torschüsse[i].lstrip().rstrip()[3:-end]
    rnk = int(Torschüsse[i][0:2].rstrip())
    data_dict[team]['Torschüsse'] = rnk
    
    # Laufdistanz:
    end = Laufdistanz[i].rstrip()[::-1].index(' ')
    end += 1
    team = Laufdistanz[i].lstrip().rstrip()[3:-end]
    rnk = int(Laufdistanz[i][0:2].rstrip())
    data_dict[team]['Laufdistanz'] = rnk
    
    # Zweikämpfe:
    end = Zweikämpfe[i].rstrip()[::-1].index(' ')
    end += 1
    team = Zweikämpfe[i].lstrip().rstrip()[3:-end]
    rnk = int(Zweikämpfe[i][0:2].rstrip())
    data_dict[team]['Zweikämpfe'] = rnk
    
    # IntensivLauf:
    end = IntensivLauf[i].rstrip()[::-1].index(' ')
    end += 1
    team = IntensivLauf[i].lstrip().rstrip()[3:-end]
    rnk = int(IntensivLauf[i][0:2].rstrip())
    data_dict[team]['IntensivLauf'] = rnk
    
    # Eckbälle:
    ind1 = Eckbälle[i].index('\t')
    ind2 = Eckbälle[i].index('\t', ind1 + 1)
    team = Eckbälle[i][ind1 + 1:ind2 ]
    Eckbälle[i] = re.sub('\t', ' ', Eckbälle[i])
    rnk = int(Eckbälle[i][0:2].rstrip())
    data_dict[team]['Eckbälle'] = rnk
```

```python
for i in range(len(Passquote)):
    ind = Tab[i][2:].lstrip().index('  ')
    team = Tab[i][2:].lstrip()[:ind]
    games = int(Tab[i][2:].lstrip()[ind+2:ind+4].rstrip())
    margin = int(Tab[i][-3:])
    data_dict[team]['margin'] = margin/games
    
    ind = TabHeim[i][2:].lstrip().index('  ')
    team = TabHeim[i][2:].lstrip()[:ind]
    games = int(TabHeim[i][2:].lstrip()[ind+2:ind+4].rstrip())
    margin = int(TabHeim[i][-3:])
    data_dict[team]['marginHeim'] = margin/games
    
    ind = TabAusw[i][2:].lstrip().index('  ')
    team = TabAusw[i][2:].lstrip()[:ind]
    games = int(TabAusw[i][2:].lstrip()[ind+2:ind+4].rstrip())
    margin = int(TabAusw[i][-3:])
    data_dict[team]['marginAusw'] = margin/games
    
    ind = TabForm[i][2:].lstrip().index('  ')
    team = TabForm[i][2:].lstrip()[:ind]
    games = int(TabForm[i][2:].lstrip()[ind+2:ind+4].rstrip())
    margin = int(TabForm[i][-3:])
    data_dict[team]['marginForm'] = margin/games
    
data_dict
```

```python
margin_ar = []
marginHeim_ar = []
marginAusw_ar = []
marginForm_ar = []
stats_ar = []
means = np.zeros((18, 2))

i = 0
for team in team_list:
    margin_ar.append([data_dict[team]['margin'], team]) 
    marginHeim_ar.append([data_dict[team]['marginHeim'], team])
    marginAusw_ar.append([data_dict[team]['marginAusw'], team])
    marginForm_ar.append([data_dict[team]['marginForm'], team])
    rnk_1 = data_dict[team]['Passquote']
    rnk_2 = data_dict[team]['Torschüsse']
    rnk_3 = data_dict[team]['Zweikämpfe']
    rnk_4 = data_dict[team]['IntensivLauf']
    rnk_5 = data_dict[team]['Laufdistanz']
    rnk_6 = data_dict[team]['Eckbälle']
    means[i, 0] = data_dict[team]['marginHeim']
    means[i, 1] = data_dict[team]['marginAusw']
    i +=1
    
    # divided by 3.5 to get the margin:
    stats_val = np.mean([rnk_1, rnk_2, rnk_3, rnk_4, rnk_5, rnk_6])/2.5
    stats_ar.append([stats_val, team, rnk_1, rnk_2, rnk_3, rnk_4, rnk_5, rnk_6])
    data_dict[team]['marginStats'] = stats_val
    
margin_rankings = sorted(margin_ar, reverse=True)
marginHeim_rankings = sorted(marginHeim_ar, reverse=True)
marginAusw_rankings = sorted(marginAusw_ar, reverse=True)
marginForm_rankings = sorted(marginForm_ar, reverse=True)
stats_rankings = sorted(stats_ar, reverse=False)

print('meanHeim: ', np.mean(means[:, 0]))
print('meanAusw: ', np.mean(means[:, 1]))
print()

print('margin:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), margin_rankings[i][1].rjust(30), '{0:+5.1f}'.format(margin_rankings[i][0] ))

print()
print('marginHeim:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), marginHeim_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginHeim_rankings[i][0] ))

print()
print('marginAusw:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), marginAusw_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginAusw_rankings[i][0] ))

print()
print('marginForm:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), marginForm_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginForm_rankings[i][0] ))
     
print()
print('marginStats:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), stats_rankings[i][1].rjust(30), '{0:5.1f}'.format(stats_rankings[i][0]), '  {}'.format(stats_rankings[i][2:] ))
```
