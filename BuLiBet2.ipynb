```python
week = 15
lastweek = week - 1
wk = str(week)
lwk = str(lastweek)

with open('Woche' + lwk + '/Spielplan_w' + wk + '.txt') as f:
    #passer = f.readlines()
    Spielplan = list(f)

with open('Woche' + lwk + '/Tabelle_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Tabelle = list(f)
    
with open('Woche' + lwk + '/TabelleHeim_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    TabelleHeim = list(f)
    
with open('Woche' + lwk + '/TabelleAusw_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    TabelleAusw = list(f)
    
with open('Woche' + lwk + '/Torschüsse_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Torschüsse = list(f)
    
with open('Woche' + lwk + '/Passquote_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Passquote = list(f)
    
with open('Woche' + lwk + '/Laufdistanz_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Laufdistanz = list(f)
    
with open('Woche' + lwk + '/Zweikämpfe_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Zweikämpfe = list(f)
    
with open('Woche' + lwk + '/Eckbälle_w' + lwk + '.txt') as f:
    #passer = f.readlines()
    Eckbälle = list(f)
```

```python
team_list = []
data_dict = {}

for i in range(len(Passquote)):
    
    # Passquote:
    j1 = Passquote[i].index('\t')
    j2 = Passquote[i].index('\t', j1 + 1)
    team = Passquote[i].lstrip().rstrip()[j1 + 1:j2]
    team_list.append(team)
    
    rnk = int(Passquote[i][:j1].rstrip())
    
    data_dict[team] = {}
    data_dict[team]['Passquote'] = rnk
    
team_list = sorted(team_list)

for i in range(len(Passquote)):
    
    # Torschüsse:
    j1 = Torschüsse[i].index('\t')
    j2 = Torschüsse[i].index('\t', j1 + 1)
    team = Torschüsse[i].lstrip().rstrip()[j1 + 1:j2]
    rnk = int(Torschüsse[i][:j1].rstrip())
    data_dict[team]['Torschüsse'] = rnk
    
    # Laufdistanz:
    j1 = Laufdistanz[i].index('\t')
    j2 = Laufdistanz[i].index('\t', j1 + 1)
    team = Laufdistanz[i].lstrip().rstrip()[j1 + 1:j2]
    rnk = int(Laufdistanz[i][:j1].rstrip())
    data_dict[team]['Laufdistanz'] = rnk
    
    # Zweikämpfe:
    j1 = Zweikämpfe[i].index('\t')
    j2 = Zweikämpfe[i].index('\t', j1 + 1)
    team = Zweikämpfe[i].lstrip().rstrip()[j1 + 1:j2]
    rnk = int(Zweikämpfe[i][:j1].rstrip())
    data_dict[team]['Zweikämpfe'] = rnk
    
    # Eckbälle:
    j1 = Eckbälle[i].index('\t')
    j2 = Eckbälle[i].index('\t', j1 + 1)
    team = Eckbälle[i].lstrip().rstrip()[j1 + 1:j2]
    rnk = int(Eckbälle[i][:j1].rstrip())
    data_dict[team]['Eckbälle'] = rnk
```

```python
for i in range(len(Passquote)):
    j1 = Tabelle[i].index('\t')
    j2 = Tabelle[i].index('\t', j1 + 1)
    team = Tabelle[i].lstrip().rstrip()[j1 + 1:j2]
    j1 = Tabelle[i].index('\t', j2 + 1)
    j2 = Tabelle[i].index('\t', j1 + 1) 
    games = int(Tabelle[i].lstrip()[j1 + 1:j2].rstrip())
    j1 = Tabelle[i][::-1].index('\t')
    j2 = Tabelle[i][::-1].index('\t', j1 + 1) 
    margin = int(Tabelle[i][::-1][j1 + 1:j2][::-1].rstrip())
    data_dict[team]['margin'] = margin/games
    
    j1 = TabelleHeim[i].index('\t')
    j2 = TabelleHeim[i].index('\t', j1 + 1)
    team = TabelleHeim[i].lstrip().rstrip()[j1 + 1:j2]
    j1 = TabelleHeim[i].index('\t', j2 + 1)
    j2 = TabelleHeim[i].index('\t', j1 + 1) 
    games = int(TabelleHeim[i].lstrip()[j1 + 1:j2].rstrip())
    j1 = TabelleHeim[i][::-1].index('\t')
    j2 = TabelleHeim[i][::-1].index('\t', j1 + 1) 
    margin = int(TabelleHeim[i][::-1][j1 + 1:j2][::-1].rstrip())
    data_dict[team]['marginHeim'] = margin/games
    
    j1 = TabelleAusw[i].index('\t')
    j2 = TabelleAusw[i].index('\t', j1 + 1)
    team = TabelleAusw[i].lstrip().rstrip()[j1 + 1:j2]
    j1 = TabelleAusw[i].index('\t', j2 + 1)
    j2 = TabelleAusw[i].index('\t', j1 + 1) 
    games = int(TabelleAusw[i].lstrip()[j1 + 1:j2].rstrip())
    j1 = TabelleAusw[i][::-1].index('\t')
    j2 = TabelleAusw[i][::-1].index('\t', j1 + 1) 
    margin = int(TabelleAusw[i][::-1][j1 + 1:j2][::-1].rstrip())
    data_dict[team]['marginAusw'] = margin/games
    
data_dict
```

```python
margin_ar = []
marginHeim_ar = []
marginAusw_ar = []
stats_ar = []
means = np.zeros((18, 2))

i = 0
for team in team_list:
    margin_ar.append([data_dict[team]['margin'], team]) 
    marginHeim_ar.append([data_dict[team]['marginHeim'], team])
    marginAusw_ar.append([data_dict[team]['marginAusw'], team])
    rnk_1 = data_dict[team]['Passquote']
    rnk_2 = data_dict[team]['Torschüsse']
    rnk_3 = data_dict[team]['Zweikämpfe']
    rnk_4 = data_dict[team]['Laufdistanz']
    rnk_5 = data_dict[team]['Eckbälle']
    means[i, 0] = data_dict[team]['marginHeim']
    means[i, 1] = data_dict[team]['marginAusw']
    i +=1
    
    # divided by 3.5 to get the margin:
    stats_val = np.mean([rnk_1, rnk_2, rnk_3, rnk_4, rnk_5])/3.5
    stats_ar.append([stats_val, team, rnk_1, rnk_2, rnk_3, rnk_4, rnk_5])
    data_dict[team]['marginStats'] = stats_val
    
margin_rankings = sorted(margin_ar, reverse=True)
marginHeim_rankings = sorted(marginHeim_ar, reverse=True)
marginAusw_rankings = sorted(marginAusw_ar, reverse=True)
stats_rankings = sorted(stats_ar, reverse=False)

meanHeim = np.mean(means[:, 0])
meanAusw = np.mean(means[:, 1])
print('meanHeim: ', meanHeim)
print('meanAusw: ', meanAusw)
print()

print('margin:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), margin_rankings[i][1].rjust(30), '{0:+5.1f}'.format(margin_rankings[i][0] ))

print()
print('marginHeim:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), marginHeim_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginHeim_rankings[i][0] ))

print()
print('marginAusw:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), marginAusw_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginAusw_rankings[i][0] ))

print()
print('marginStats:')
print()
for i in range(18):
    print('{:2d}'.format(i + 1), stats_rankings[i][1].rjust(30), '{0:5.1f}'.format(stats_rankings[i][0]), '  {}'.format(stats_rankings[i][2:] ))
```
