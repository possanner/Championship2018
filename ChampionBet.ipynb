```python
for i in range(2, 26):
    
    Goals[i] = re.sub('Birmingham City', 'Birmingham', Goals[i])
    Goals[i] = re.sub('Derby County', 'Derby', Goals[i])
    Goals[i] = re.sub('Hull City', 'Hull', Goals[i])
    Goals[i] = re.sub('Ipswich Town', 'Ipswich', Goals[i])
    Goals[i] = re.sub('Leeds Utd', 'Leeds', Goals[i])
    Goals[i] = re.sub('Norwich City', 'Norwich', Goals[i])
    Goals[i] = re.sub('Nottm Forest', 'Nottingham Forest', Goals[i])
    Goals[i] = re.sub('QP Rangers', 'Queens Park Rangers', Goals[i])
    Goals[i] = re.sub('Sheffield Utd', 'Sheffield United', Goals[i])
    Goals[i] = re.sub('Sheffield Wed', 'Sheffield Wednesday', Goals[i])
    Goals[i] = re.sub('Stoke City', 'Stoke', Goals[i])
    Goals[i] = re.sub('Swansea City', 'Swansea', Goals[i])
    Goals[i] = re.sub('West Bromwich', 'West Bromwich Albion', Goals[i])
    Goals[i] = re.sub('Wigan Athletic', 'Wigan', Goals[i])
    
for i in range(2, 26):
    
    j1 = Goals[i].index('\t')
    team = Goals[i][:j1].lstrip().rstrip()
    print(team)
    j2 = Goals[i].index('\t', j1 + 1) 
    scored_home = float(Goals[i].lstrip()[j1:j2].rstrip())
    j3 = Goals[i].index('\t', j2 + 1) 
    conceded_home = float(Goals[i].lstrip()[j2:j3].rstrip())
    j4 = Goals[i].index('\t', j3 + 1)
    j5 = Goals[i].index('\t', j4 + 1)
    j6 = Goals[i].index('\t', j5 + 1)
    scored_away = float(Goals[i].lstrip()[j5:j6].rstrip())
    j7 = Goals[i].index('\t', j6 + 1)
    conceded_away = float(Goals[i].lstrip()[j6:j7].rstrip())
    j8 = Goals[i].index('\t', j7 + 1)
    j9 = Goals[i].index('\t', j8 + 1)
    j10 = Goals[i].index('\t', j9 + 1)
    scored_tot = float(Goals[i].lstrip()[j9:j10].rstrip())
    j11 = Goals[i].index('\t', j10 + 1)
    conceded_tot = float(Goals[i].lstrip()[j10:j11].rstrip())
    
    margin_home = scored_home - conceded_home
    margin_away = scored_away - conceded_away
    margin_tot = scored_tot - conceded_tot
    
    data_dict[team]['margin home'] = margin_home
    data_dict[team]['margin away'] = margin_away
    data_dict[team]['margin tot'] = margin_tot
```

```python
margin_ar = []
marginHome_ar = []
marginAway_ar = []
marginForm_ar = []
stats_ar = []
means = np.zeros((24, 2))

i = 0
for team in team_list:
    margin_ar.append([data_dict[team]['margin tot'], team]) 
    marginHome_ar.append([data_dict[team]['margin home'], team])
    marginAway_ar.append([data_dict[team]['margin away'], team])
    marginForm_ar.append([data_dict[team]['margin form'], team])
    #rnk_1 = data_dict[team]['Corners']
    rnk_2 = data_dict[team]['Passes']
    rnk_3 = data_dict[team]['Shots']
    #rnk_4 = data_dict[team]['BigChances']
    means[i, 0] = data_dict[team]['margin home']
    means[i, 1] = data_dict[team]['margin away']
    i +=1
    
    # divided by 6.0 to get the margin:
    stats_val = np.mean([rnk_2, rnk_3])/6.5
    stats_ar.append([stats_val, team, rnk_2, rnk_3])
    data_dict[team]['margin stats'] = stats_val
    
margin_rankings = sorted(margin_ar, reverse=True)
marginHome_rankings = sorted(marginHome_ar, reverse=True)
marginAway_rankings = sorted(marginAway_ar, reverse=True)
marginForm_rankings = sorted(marginForm_ar, reverse=True)
stats_rankings = sorted(stats_ar, reverse=False)

meanHome = np.mean(means[:, 0])
meanAway = np.mean(means[:, 1])
print('meanHome: ', meanHome)
print('meanAway: ', meanAway)
print()

print('margin:')
print()
for i in range(24):
    print('{:2d}'.format(i + 1), margin_rankings[i][1].rjust(30), '{0:+5.1f}'.format(margin_rankings[i][0] ))

print()
print('marginHome:')
print()
for i in range(24):
    print('{:2d}'.format(i + 1), marginHome_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginHome_rankings[i][0] ))

print()
print('marginAway:')
print()
for i in range(24):
    print('{:2d}'.format(i + 1), marginAway_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginAway_rankings[i][0] ))

print()
print('marginForm:')
print()
for i in range(24):
    print('{:2d}'.format(i + 1), marginForm_rankings[i][1].rjust(30), '{0:+5.1f}'.format(marginForm_rankings[i][0] ))
    
print()
print('marginStats:')
print()
for i in range(24):
    print('{:2d}'.format(i + 1), stats_rankings[i][1].rjust(30), '{0:5.1f}'.format(stats_rankings[i][0]), '  {}'.format(stats_rankings[i][2:] ))
```
